# Task 1 - Extracting the frequency distribution

Your goal is to extract the frequency distribution of all the characters present in the text. The characters are all letters from `a` to `z` including `spaces` and `.`. So that's a total of 28 characters in all.

You need to fill in the function `extract_freq_dist(text)` which takes in the text (as a string) and outputs a dictionary with the frequency of each letter. We'll do our Huffman coding in frequency form to avoid floating point errors!



```python
###############################################
# SANDBOX FOR TESTING
###############################################

freq = extract_freq_dist(hidden_message)

print("Number of characters:", len(freq))
print("Number of spaces:", freq[" "])
print("Number of s:", freq["s"])
print("Number of z:", freq["z"])




###############################################
```

```python
###############################################
# Frequency extraction function
###############################################
def extract_freq_dist(text):
  # Fill in your code here
  placeholder = {}
  ctr = 0
  textlength = len(text)
  while ctr < textlength:
    char = text[ctr]
    if char in placeholder: placeholder[char] += 1
    else: placeholder[char] = 1
    ctr += 1

  return placeholder
```

# Task 2 - Huffman coding function

This is the exciting part. Fill in the Huffman coding function `huffman_codebook(freq_dict)`. It takes in the frequency distribution you just created and returns another dictionary showing the codebook. Just a reminder, a codebook just shows the mapping of the source symbols to the Huffman codewords.

THIS TABLE IS JUST AN EXAMPLE. IT'S NOT THE CORRECT ONE.

Symbol             | Codewords        | 
-------------------|------------------|
a                  |  1011            |
b                  |  10              |
c                  |  10110           |
d                  |  100             |

The output dictionary should have something like:

`{'a':'1011', 'b':'10', 'c':'10110', 'd':'100'}`

Where the binary code are in strings. We'll use these for decoding hidden messages later.

A few guidelines:
* Make sure the Huffman coding follows the discussion from our Wiki page
* All left branches are tagged with 0s while all right branches are tagged with 1s
* In case of equal probabilities, the deeper tree takes the left path. If they have the same depth, then sort alphabetically
* Actually... Python dictionaries already sort keys first by string length, then by alphabetical order. So you wouldn't worry too much



```python
###############################################
# Huffman coding function
###############################################
def huffman_codebook(freq_dict):
  # Fill in your code here
  codebook = dict.fromkeys(freq_dict.keys(),'')

  while len(freq_dict) != 1:
    freq_dict = dict(sorted(freq_dict.items(), key=lambda item: item[1])) #sort dictionary in ascending order
    symbols = list(freq_dict.keys()) #create sorted symbols
    freq = list(freq_dict.values()) #create sorted frequencies

    freq_dict[symbols[0]+symbols[1]] = freq[0]+freq[1] #create new item in dictionary

    # left side of tree
    for i in codebook.keys(): #iterate through symbols
      if i in symbols[0]: #check if character is in the left
        codebook[i] = "0" + codebook[i] #add 0

    #right side of tree
    for i in codebook.keys(): #iterate through symbols
      if i in symbols[1]: #check if character is in the right
        codebook[i] = "1" + codebook[i] #add 1

    del freq_dict[symbols[0]] #delete smallest item
    del freq_dict[symbols[1]] #delete 2nd smallest item
    symbols = [] #reset placeholder
    freq = [] #reset placeholder

  return codebook


##############################################
# test
freq_dict = extract_freq_dist(hidden_message)
codebook = huffman_codebook(freq_dict)

# print(freq_dict)
# freq_dict = dict(sorted(freq_dict.items(), key=lambda item: item[1]))
# print(freq_dict)
# symbols = list(freq_dict.keys())
# freq = list(freq_dict.values())
# freq_dict[symbols[0]+symbols[1]] = freq[0]+freq[1]
# del freq_dict[symbols[0]]
# del freq_dict[symbols[1]]
# print(freq_dict)
# print()

# for key in codebook.keys():
#   codebook[key] = "0" + codebook[key]
# print(codebook)
# for key in codebook.keys():
#   codebook[key] = "1" + codebook[key]
# print(codebook)

print(codebook)
```

#Task 3 - Decode the hidden messages

In this task you'll need to create a function `decode_message(decodebook,text)`. 
* The decodebook is simply the swapped keys and values of our codebook. That means all binary values are now keys while all the keys are now the values. We'll do this for you.
*The text argument is a string of binary symbols (e.g.,`1110101001`). The function does not return anything except that it prints the decoded message.



```python
###############################################
# DO NOT TOUCH. This is the dictionary swapping code.
# Yari ka na pag ginalaw mo ito >:(
###############################################

# Swap key-value terms
decodebook = dict(map(reversed,codebook.items()))

# Check me here
print(decodebook)
```

```python
###############################################
# Decoding function
###############################################
def decode_message(decodebook,text):
  # Fill in your code here
  message = ''
  check = ''
  ctr = 0
  while ctr < len(text):
    check = check + text[ctr]
    if check in decodebook.keys():
      message += decodebook[check]
      check = ''
    ctr += 1

  return message
```
